```python
data = data[data["size"].notna()]
```

```python
data["bhk"] = data["size"].apply(lambda x: int(x.split(" ")[0]))
```

```python
def converRangeToNum(x):
    token = x.split("-")
    if len(token) ==  2:
        return ((float(token[0]) + float(token[1]))/2)
    try:
        return float(x)
    except:
        None
```

```python
data["total_sqft"] = data["total_sqft"].apply(converRangeToNum)
```

```python
data.total_sqft.unique()
```

```python
data.describe()
```

```python
dataclean = data.copy()
```

```python
dataclean
```

```python
dataclean["price_per_sq"] = dataclean["price"]*100000/dataclean["total_sqft"]
```

```python
dataclean
```

```python
location_stats = dataclean.groupby("location")["location"].agg("count").sort_values(ascending = False)
```

```python
location_stats
```

```python
location_less_then_10 = location_stats[location_stats <=10]
```

```python
location_less_then_10
```

```python
dataclean["location"] = dataclean["location"].apply(lambda x : "other" if (x in location_less_then_10) else x)
```

```python
dataclean.location
```

```python
dataclean.head(15)
```

```python
dataclean = dataclean[~((dataclean.total_sqft / dataclean.bhk )<300)]
```

```python
dataclean
```

```python
data0 = dataclean.copy()
data0.location = data0.location.apply(lambda x : x.strip())
```

```python
dataclean.shape
```

```python
def remove_outliers(df):
    out_df = pd.DataFrame()
    for k,subdf in df.groupby("location"):
        m = np.mean(subdf.price_per_sq)
        s = np.std(subdf.price_per_sq)
        red_df = subdf[(subdf.price_per_sq > (m-s)) &(subdf.price_per_sq <=(m+s))] 
        out_df = pd.concat([out_df,red_df],ignore_index=True)
    return out_df
```

```python
dataclean1 = remove_outliers(data0)
```

```python
dataclean1.head()
```

```python
dataclean1.shape
```

```python
dataclean2 = dataclean1[dataclean1.bath<dataclean1.bhk+2]
```

```python
dataclean2.shape
```

```python
dataclean2.head(3)
```

```python
data1 = dataclean2.drop(columns=["area_type","availability","society","balcony","price_per_sq"])
```

```python
data1.head(10)
```

```python
dummies = pd.get_dummies(data1.location)
```

```python
data1.head()
```

```python
dummies.head(5)
```

```python
data2 = pd.concat([data1,dummies.drop(columns="other")],axis="columns")
```

```python
data2
```

```python
data2.location.unique()
```

```python
data3 = data2.drop(columns="location")
```

```python
data3.head()
```

```python
X = data3.drop(columns="price")
```

```python
X
```

```python
y = data3.price
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,train_size = 0.2,random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
obj = LinearRegression()
obj.fit(X_train,y_train)
obj.score(X_test,y_test)
```

```python
X.info()
```

```python
np.where(X.columns == "1st Block Jayanagar")[0][0]
```

```python
def predict_price(location,total_sq,bath,bhk):
    loc_indx = np.where(X.columns == location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] = total_sq
    x[1] = bath
    x[2] = bhk
    
    if loc_indx >= 0:
        x[loc_indx] = 1
    return obj.predict([x])[0]
        
```

```python
predict_price("1st Block Jayanagar",1000,2,3)
```
